```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate data points to plot 
#data should be of same size
x=[x for x in range(-1,10)]
y=[x for x in range(-5,6)]
```

```python
#ploting points of (x,y) co-ordinates using scatter() function
plt.scatter(x,y)
```

```python
#use argument 'c' to give colour for points
plt.scatter(x,y,c="y")
```

```python
plt.scatter(x,y,c="y")
#use argument 'ylim' to set desired range of yaxis
plt.ylim(-10,10)
#use argument 'xlim' to set desired range of xaxis
plt.xlim(-10,11)
```

```python
#generate random data points to plot
#data should be of same size
#np.random.normal(mean, std, number of points)
x=np.random.normal(100,5,150)
y=np.random.normal(130,5,150)
```

```python
#use argument 's' to set size of points
plt.scatter(x,y,c="y",s=0.5)
```

```python
#use argument 'label' to set legend for points
plt.scatter(x,y,c="y",s=0.9,label="random points")
#display legend using legend() 
plt.legend()
```

```python
plt.scatter(x,y,c="y",s=0.9,label="random points")
plt.legend()
#display title of graph
plt.title("scatter graph")
#display x-label of graph
plt.xlabel("x points")
#display y-label of graph
plt.ylabel("y points")
```

```python
#ploting line graph for (x,y) co-ordinates using plot() function
plt.plot(x,y,c="y",label="random points")
plt.legend()
plt.title("line graph")
plt.xlabel("x points")
plt.ylabel("y points")
```

```python
#generate points to plot sine-wave
#np.linspace(start, end, number of points)
x=np.linspace(-4*np.pi,4*np.pi,170)
#create y as array , contaning values sin(x) for each point in x
y=np.sin(x)
```

```python
#ploting line-graph of sine function
plt.plot(x,y,c="y",label="sin(x)")
plt.legend()
plt.title("sine-wave")
plt.xlabel("x points")
plt.ylabel("y points")
```

```python
#ploting scatter-graph of sine function
plt.scatter(x,y,c="y",label="sin(x)")
plt.legend()
plt.title("sine-wave")
plt.xlabel("x points")
plt.ylabel("y points")
```

```python
#create y2 as array , contaning values cos(x) for each point in x
y2=np.cos(x)
```

```python
#ploting scatter-graph of cosine function
plt.scatter(x,y2,c="y",label="cos(x)")
plt.legend()
plt.title("cosine-wave")
plt.xlabel("x points")
plt.ylabel("y points")
```

```python
#ploting line-graph of cosine function
plt.plot(x,y2,c="y",label="cos(x)")
plt.legend()
plt.title("cosine-wave")
plt.xlabel("x points")
plt.ylabel("y points")
```

```python
dart=np.random.normal(70,1,250)
```

```python
#ploting histogram for dart 
plt.hist(dart,label="counts")
plt.legend()
plt.title("histogram")
plt.xlabel("data")
plt.ylabel("value")
```

```python
#use argument 'bins' to set number of bins
plt.hist(dart,label="counts",bins=5)
plt.legend()
plt.title("histogram")
plt.xlabel("data")
plt.ylabel("value")
```

```python
#use argument 'width' to set width of bins
plt.hist(dart,label="counts",bins=15,width=0.3)
plt.legend()
plt.title("histogram")
plt.xlabel("data")
plt.ylabel("value")
```

```python
xp=["a","b","c","d","e"]
```

```python
yp=[56,67,87,90,10]
```

```python
#poting piechart
plt.pie(yp, labels=xp)
plt.legend()
plt.title("piechart")
```

```python
#use argument 'autopct' to auto generate percentage
plt.pie(yp, labels=xp, autopct='%.2f%%')
plt.legend()
plt.title("piechart")
```

```python
ex=[0,0,0,0,0.2]
```

```python
#use argument 'explode' to explode parts of piechart
plt.pie(yp, labels=xp, autopct='%.2f%%', explode=ex)
plt.legend()
plt.title("piechart")
```
